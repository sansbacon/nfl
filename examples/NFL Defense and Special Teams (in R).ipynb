# Defense and Special Teams (DST)

This is an analysis of DST scoring that doubles as a cookbook for R data manipulation and visualization techniques.

## Setting up the R Environment

I've created a small R package called "rnfl" that I use to access the relevant datasets.

```r
suppressMessages(library(rnfl))
con <- dbcon()
```

## DST Scoring

We'll start with dataset.dst, which returns a dataframe with the DST weekly scoring results since 2009. It also includes columns for the vegas total, team implied totals, win/loss, and home/away, which we will use later on to test some theories.

```r
# get data and create variable for plays per game
dst = dataset.dst(con)
str(dst)
```

We'll call the dst.factors function from rnfl, which makes some of the columns "factor" variables. This will make it easier to analyze DST scoring by category or group, for example, in wins vs. losses or home vs. away games.

```r
str(dst.factors(dst))
```

### Components of DST Scoring

We'll call the dst.scoring function from rnfl, which uses the following values for DST scoring:

* Sacks: 1 point
* Interceptions: 2 points
* Fumbles Recovered: 2 points
* Safeties: 2 points
* DST TD: 6 points
* Points Allowed 
  * (0): 10 points
  * (1-6): 7 points
  * (7-13): 4 points
  * (14-20): 1 points
  * (21-27): 0 points
  * (28-34): -1 points
  * (35+): -4 points
  
The dataframe includes not only the total DST points, but also the points for each subcategory that makes up DST scoring, such as sacks, interceptions, points allowed, etc.

```r
str(dst.scoring.std(dst))
```

The dst.scoring.parts function lets us take a look at how the average team for each weekly rank scores its points. The first table below shows that the highest scoring teams tend to derive more value from defensive and special teams touchdowns, which makes sense. You can also see that scoring from points allowed does not vary nearly as much as sacks and touchdowns.

```r
head(dst.scoring.parts(dst), 10)
```

This table groups weekly_rank into 8 subgroups (1 is 1-4, 2 is 5-8, etc.) and performs the same analysis. We can see even more clearly that the difference between the very top and the above-average defenses is defensive and special teams touchdowns.

```r
dst.scoring.parts(dst, groups=8)
```

This naturally leads us to ask how defenses score touchdowns and assess whether we can come up with a predictive model for touchdown scoring. A starting point is to look at the correlations between variables in each group.

```r
dst.scoring.std(dst) %>%
  rename(fpts = fantasy_points, ptsa = pts_allowed) %>%
  select(fpts, ptsa, sacks, int, fum_rec, def_td, return_td, safeties) %>%
  cor() %>% round(., 3)
```

We can take visualize the correlation matrix using the corrplot library. The plots tend to be a bit large, so we can set the height and width ahead of time to make them more compact.

```r
options(repr.plot.width=5, repr.plot.height=3)
library(corrplot)
dst.scoring.std(dst) %>%
  rename(fpts = fantasy_points, ptsa = pts_allowed, fum = fum_rec, dtd = def_td, sttd = return_td, sft = safeties) %>%
  select(fpts, ptsa, sacks, int, fum, dtd, sttd, sft) %>%
  cor() %>% corrplot()
```

For teams with a weekly rank between 1 and 8, most of the correlations fade away, but defensive touchdowns still have a strong correlation with fantasy points scored. I think this suggests that these teams are having roughly the same amount of sacks and giving up the same amount of points, and the main differentiator is TD scoring.

```r
dst.scoring.std(dst) %>%
  rename(fpts = fantasy_points, ptsa = pts_allowed) %>%
  filter(weekly_rank <= 8) %>%
  select(fpts, ptsa, sacks, int, fum_rec, def_td, return_td, safeties) %>%
  cor() %>% round(., 3)
```

```r
dst.scoring.std(dst) %>%
  rename(fpts = fantasy_points, ptsa = pts_allowed) %>%
  filter(weekly_rank <= 8) %>%
  select(fpts, ptsa, sacks, int, fum_rec, def_td, return_td, safeties) %>%
  cor() %>% corrplot()
```

## Looking at Home/Road and Favorite/Underdog Splits

Explain this section . . . 

```r
seasons = max(dst$season_year) - min(dst$season_year) + 1
weeks = max(dst$week) - min(dst$week) + 1
```

```r
dst.scoring.std(dst) %>%
  group_by(season_year, is_home) %>%
  summarize(n=n(), 
            avgrk=round(mean(weekly_rank), 1), 
            avgpts=round(mean(fantasy_points), 1),
            top3=round(sum(weekly_rank <= 3)/(3*weeks), 2),
            top8=round(sum(weekly_rank <= 8)/(8*weeks), 2),
            top16=round(sum(weekly_rank <= 16)/(16*weeks), 2)) %>%
  arrange(desc(season_year), desc(is_home)) %>% head(, 6)
```

```r
dst.scoring.std(dst) %>%
  mutate(is_favorite = as.factor(ifelse(impltot > opp_impltot, TRUE, FALSE))) %>%
  mutate(is_big_favorite = as.factor(ifelse(impltot - opp_impltot >= 7, TRUE, FALSE))) %>%
  mutate(is_huge_favorite = as.factor(ifelse(impltot - opp_impltot >= 10, TRUE, FALSE))) %>%
  mutate(is_big_dog = as.factor(ifelse(impltot - opp_impltot <= -7, TRUE, FALSE))) %>%
  group_by(is_home, is_favorite, is_big_favorite, is_huge_favorite, is_big_dog) %>%
  summarize(n=n(), nperc=round(n()/nrow(.),2),
            avgrk=round(mean(weekly_rank), 1), 
            avgpts=round(mean(fantasy_points), 1),
            top3=round(sum(weekly_rank <= 3)/(3*weeks*seasons), 2),
            top8=round(sum(weekly_rank <= 8)/(8*weeks*seasons), 2),
            top16=round(sum(weekly_rank <= 16)/(16*weeks*seasons), 2)
  ) %>%
  mutate(top3d = top3 - nperc, top8d = top8 - nperc, top16d = top16 - nperc) %>%
  arrange(desc(is_home), desc(is_huge_favorite), desc(is_big_favorite), desc(is_favorite))
```

# Reviewing the 2016 Season

```r
# summary of points scored by DST of team_code
dst %>%
  filter(season_year == 2016) %>%
  group_by(team_code) %>%
  summarize(min_pts = min(fantasy_points),
          p25_pts = round(quantile(fantasy_points, .25),1), 
          med_pts = median(fantasy_points), 
          p75_pts = round(quantile(fantasy_points, .75),1),
          max_pts = max(fantasy_points)) %>%
  arrange(desc(p25_pts)) %>%
  head(., 10)
```

```r
# group summary of points scored by DST of team_code
dst %>%
  group_by(season_year, team_code) %>%
  summarize(lt5 = sum(fantasy_points < 5),
      lt10 = sum(fantasy_points < 10 & fantasy_points >= 5),
          bw1020 = sum(fantasy_points >= 10 & fantasy_points <= 20), 
          gt20 = sum(fantasy_points > 20)) %>%
  arrange(desc(bw1020)) %>%
  head(., 10)
```

```r
# top N / bottom N finishes last two years
dst %>%
  filter(season_year == 2016) %>%
  group_by(team_code) %>%
  summarize(top3=sum(weekly_rank<=3), 
            top5=sum(weekly_rank<=5), 
            top10=sum(weekly_rank<=10),
            tophalf=sum(weekly_rank<=16),
            bot3=sum(weekly_rank>=30),
            bot5=sum(weekly_rank>=28),
            bot10=sum(weekly_rank>=23),
            bothalf=sum(weekly_rank>=17)) %>%
  arrange(desc(tophalf)) %>%
  head(., 10)
```

## Visualizing DST Scoring

```r
# use group_by to calculate summary statistics by season 
dst.summ =  dst %>% 
  group_by(season_year, week) %>% 
  summarize(min_pts = round(min(fantasy_points),1),
          med_pts = round(median(fantasy_points),1), 
          p25_pts = round(quantile(fantasy_points, .25),1), 
          p75_pts = round(quantile(fantasy_points, .75),1),
          p90_pts = round(quantile(fantasy_points, .9),1),
          max_pts = round(max(fantasy_points),1)) %>%
  group_by(season_year) %>%
  summarize(min_pts = round(mean(min_pts),1),
          med_pts = round(mean(med_pts),1), 
          p25_pts = round(mean(p25_pts),1), 
          p75_pts = round(mean(p75_pts),1),
          p90_pts = round(mean(p90_pts),1),
          max_pts = round(mean(max_pts),1))
head(dst.summ)
```

### Line Plot

There are two different ways to compose a line plot, as covered previously. We will "melt" the data using the reshape2 package. Your columns will now be season_year, variable, and value. You can then plot season_year on the x-axis and value on the y-axis, and use the variable column for group and color. The values are the weekly average per season for the max, 90th percentile, 75th percentile, median, 25th percentile, and minimum points scored.

```r
# melt the data into season_year, variable, value
# each variable will be a factor (min_pts, etc.)
melt(dst.summ, id.vars = c("season_year")) %>%
ggplot(aes(x=season_year, y=value)) + 
  geom_line(aes(color=variable, group=variable)) +
  labs(title = "DST Fantasy Points: 2009 - 2016", y = "Avg DST Points", x="Season") +
  theme(plot.title = element_text(hjust = 0.5))
```
