# NFL Offense

This is an overview of trends in NFL offense that doubles as a cookbook for R data manipulation and visualization techniques. We'll look at team plays, team touchdowns, team passing, and team rushing. This provides necessary context for evaluating opportunity for individual players.

## Setting up the R Environment

```r
library(dplyr)
library(ggplot2)
library(reshape2)
options(repr.plot.width=6, repr.plot.height=3.5)
```

## Setting up the Database Connection

I have my NFL data stored in a postgresql database. I started by using burntsushi's nfldb python library, which draws from the play-by-play data that powers the NFL.com Gamecenter. I then wrote some queries that aggregate the play-by-play data into week and season-level stats. If you don't want to do this, you can get yearly team-level statistics from the statistics pages at NFL.com, ESPN.com, ProFootballReference.com, or other sites. See the Appendix for additional details.

```r
require(RPostgreSQL)
con <- dbcon()
```

# Team Plays Per Season

```r
# get data and create variable for plays per game
teamo = dataset.team.offense(con)
teamo$plays_pg = teamo$play_count_tip / 16
```

```r
# use group_by to calculate summary statistics by season 
yt =  teamo %>% group_by(season_year) %>% 
summarize(min_plpg = round(min(plays_pg),1),
          avg_plpg = round(mean(plays_pg),1),
          med_plpg = round(median(plays_pg),1), 
          p25_plpg = round(quantile(plays_pg, .25),1), 
          p75_plpg = round(quantile(plays_pg, .75),1),
          max_plpg = round(max(plays_pg),1)
)
```

## Visualizing Team Plays

### Line Plot

There are two different ways to compose a line plot. The first is to add a geom_line element for each line you want to plot. You can then use scale_color_manual to associate a color with each line. While this method is somewhat clunky in that you have to specify a color for every line, you do not have to make any changes to the underlying data.

The second way is to "melt" the data using the reshape2 package. Your columns will now be season_year, variable, and value. You can then plot season_year on the x-axis and value on the y-axis, and use the variable column for group and color.

```r
# Method #1: 
options(repr.plot.width=6, repr.plot.height=3)
cols <- c("Max"="gold", "75th Pctl"="gray64", "Median"="darkorange3", "25th Pctl"="blue", "Min"="red") 

# season_year on x-axis, value of each plays_per_game group on the y-axis
ggplot(yt, aes(x=season_year)) + 
geom_line(aes(y=max_plpg, group=1, color='Max')) + 
geom_line(aes(y=p75_plpg, group=1, color='75th Pctl')) +
geom_line(aes(y=med_plpg, group=1, color='Median')) + 
geom_line(aes(y=p25_plpg, group=1, color='25th Pctl')) +
geom_line(aes(y=min_plpg, group=1, color='Min')) +
scale_color_manual("Plays Per Game", values = cols, breaks = c("Max", "75th Pctl", "Median", "25th Pctl", "Min")) +
labs(title = "Plays Per Game: 2009 - 2016", y = "Plays Per Game", x="NFL Season") +
scale_x_continuous(breaks=seq(2009,2016,1)) +
scale_y_continuous(breaks=seq(50,80,2)) + 
theme(plot.title = element_text(hjust = 0.5), axis.title.x = element_text(hjust=.5))
```

```r
# melt the data into season_year, variable, value
yt2 = subset(yt, select=c(season_year, max_plpg, p75_plpg, med_plpg, p25_plpg, min_plpg))
yt2 = melt(yt2, id.vars = c("season_year"))
```

```r
# plot melted data, can use variable for color and group
ggplot(yt2, aes(x=season_year, y=value)) + geom_line(aes(color=variable, group=variable)) + 
labs(title = "Plays Per Game: 2009 - 2016", y = "Plays Per Game", x="Season") +
scale_x_continuous(breaks=seq(2009,2016,1)) +
scale_y_continuous(breaks=seq(50,80,2)) + 
theme(plot.title = element_text(hjust = 0.5)) +
scale_colour_discrete(name="Plays Per Game", h = c(100, 310), 
                    labels = c("Max", "75th Pctl", "Median", "25th Pctl", "Min"))
```

### Violin Plot

The violin plot below shows the concentration of team plays from 2009-2016. The lines within each plot represent the 25th, 50th, and 75th percentiles. The width and height of the plot shows how far apart values are from each other. For example, in 2011, which has a shorter and wider plot, a large number of teams had between 64-67 plays per game, whereas team averages were more spread out in the prior and following years.

```r
# need to make minor changes to dataset for violin plot
# season_year as factor will create separate violin for each one
teamo$season_year = as.factor(teamo$season_year)

# use ifelse to create 4 ranked groups of teams based on total plays 
# (all teams play same games so rank is same for total plays and plays per game)
# these are used to color the jitter plot within the violin plot
teamo$plays_grp = as.factor(ifelse(teamo$plays_rank <= 8, 1, 
                  ifelse(teamo$plays_rank <= 16, 2, 
                  ifelse(teamo$plays_rank <= 24, 3, 4))))
```

```r
# first create the violin plot with 25th, 50th, 75th percentiles
# then add jitter where color is the rank group (1-4)
vp = ggplot(teamo, aes(season_year, plays_pg)) + 
geom_violin(draw_quantiles = c(0.25, 0.5, 0.75)) +  
geom_jitter(aes(color=plays_grp), height=0, width=0.2) +
labs(title = "Plays Per Game: 2009 - 2016", y = "Plays", x="Season") + 
scale_y_continuous(breaks=seq(55,80,2)) + 
theme(plot.title = element_text(hjust = 0.5))

# violin plots need to be a little bit larger to properly display data
options(repr.plot.width=8, repr.plot.height=5)

# add a color scale that clearly distinguishes the four rank groups
vp + scale_colour_discrete(name="Team Rank", h = c(100, 310),
                         breaks=c("1", "2", "3", "4"),
                         labels=c("1-8", "9-16", "17-24", "25-32"))
```

## Database Setup

```r
dbcon <- function(conf='~/.rconfig') {
  source(conf)
  drv <- dbDriver('PostgreSQL')
  dbConnect(drv, dbname = nfldb.database, user=nfldb.username, password=nfldb.password)
}
```

```r
dataset.team.offense <- function(connxn, seas=2009) {
  q = paste0(
      "SELECT *, (pass_td + rush_td) as td, ",
      "row_number() OVER (PARTITION BY season_year ORDER BY play_count_tip DESC) as plays_rank, ",
      "row_number() OVER (PARTITION BY season_year ORDER BY (pass_td + rush_td) DESC) as td_rank ",
      "FROM teamstats_offense_yearly WHERE season_year >= ", seas, 
      " ORDER BY season_year, pass_att DESC")
  dbGetQuery(connxn, q)
}
```

```r
dataset.team.plays <- function(connxn, seas=2009) {
    q = paste0(
      "SELECT season_year, week, game_date, team_code, opp, ",
      "row_number() OVER (PARTITION BY team_code ORDER BY season_year, week) AS gamenum, ",
      "s, consensus_spread, consensus_game_ou, consensus_implied_total, is_win, plays ",
      "FROM vw_teamplays ",
      "WHERE gsis_id <> '2014122809' AND season_year >= ", seas,
      " ORDER BY gsis_id, team_code")
   dbGetQuery(connxn, q)
}
```

```r
dataset.td <- function(connxn, seas=2009) {
    q = paste0(
      "SELECT season_year, week, team, opp, phase, passing_yds, passing_tds, rushing_yds, rushing_tds ",
      "FROM offense_tds ",
      "WHERE season_year >= ", seas)
   dbGetQuery(connxn, q)
}
```

## Utility Functions

```r
# helper function to plot multiple graphs on same plot
multiplot <- function(..., plotlist=NULL, file, cols=1, layout=NULL) {
    library(grid)
    
    # Make a list from the ... arguments and plotlist
    plots <- c(list(...), plotlist)
    
    numPlots = length(plots)
    
    # If layout is NULL, then use 'cols' to determine layout
    if (is.null(layout)) {
        # Make the panel
        # ncol: Number of columns of plots
        # nrow: Number of rows needed, calculated from # of cols
        layout <- matrix(seq(1, cols * ceiling(numPlots/cols)),
                     ncol = cols, nrow = ceiling(numPlots/cols))
    }
    
    if (numPlots==1) {
        print(plots[[1]])
        
    } else {
        # Set up the page
        grid.newpage()
        pushViewport(viewport(layout = grid.layout(nrow(layout), ncol(layout))))
        
        # Make each plot, in the correct location
        for (i in 1:numPlots) {
           # Get the i,j matrix positions of the regions that contain this subplot
           matchidx <- as.data.frame(which(layout == i, arr.ind = TRUE))
           
           print(plots[[i]], vp = viewport(layout.pos.row = matchidx$row,
                                    layout.pos.col = matchidx$col))
        }
    }
}
```
