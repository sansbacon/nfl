# Introduction to Air Yards

"Receiver Air Yards are the total number of yards thrown toward a receiver on a play in which he is targeted, both complete and incomplete. If you add them up over a game or a season, you get a receiver’s total Air Yards . . . Air Yards aren’t a measure of how much distance the ball actually travelled. Instead, they measure potential yards a receiver would have produced if he caught the ball and then was immediately tackled." Josh Hermsmeyer, <a href='http://www.rotoworld.com/articles/nfl/69265/446/why-receiver-air-yards-matter'>Why Receiver Air Yards Matter</a> (Feb. 15, 2017).

As explained in this article (as well as articles and podcasts on RotoViz.com), we care about Air Yards because they (1) give us information about how an offense intends to use a receiver and (2) help identify receivers whose production could meaningfully change if their efficiency spikes or falls.

## The NFL GameCenter API

The term "API" is used loosely, as NFL.com does not have an official mechanism to make requests for NFL data and it does not document any endpoints or document the various fields in a response. So we have to use trial and error to figure out how the GameCenter app gets its data and make educated guesses about the meaning of various fields.

We'll start with a list of games per week. NFL.com has a scoreboard page for each week of the season (beginning in 2009). The URL has the following pattern:

    www.nfl.com/scores/{season_year}/REG{season_week}
    
The code below gets the scoreboard page for any week of any season (since 2009).

```python
import requests

def get_scoreboard(season, week):
    '''
    Gets scoreboard page from nfl.com
    
    Arguments:
        season: int between 2009 and present season
        week: int between 1 and 17 (if you want regular season, goes higher for playoffs)
        
    Returns:
        HTML content of webpage
    '''
    base_url = 'http://www.nfl.com/scores/{season}/REG{week}'
    r = requests.get(base_url.format(season, week))
    r.raise_for_status()
    return r.content
```

Each scoreboard page has links ('a' elements in HTML) to individual GameCenter pages. These links follow this format:

    http://www.nfl.com/gamecenter/{gsis_id}/{season}/REG{week}/{away}@{home}

Here is the link for the Chiefs-Chargers game from week 1 of the 2016 NFL season:

    http://www.nfl.com/gamecenter/2016091104/2016/REG1/chargers@chiefs

Using your browser's developer tools, you can see that the GameCenter links all have the CSS class 'game-center-link'. 

    <a href="/gamecenter/2016091104/2016/REG1/chargers@chiefs" class="game-center-link" ... >

The code below uses this information to find all of the GameCenter links on a scoreboard page.

```python
from bs4 import BeautifulSoup


def get_gamecenter_links(page):
    '''
    Gets all gamecenter links from a scoreboard page
    
    Arguments:
        page: str of HTML source
        
    Returns:
        list of URLs
    '''
    soup = BeautifulSoup(page, 'lxml')
    return [url['href'] for url in soup.find_all('a', {'class': 'game-center-link'})]
```

You can then use your browser's developer tools to inspect the network activity when you load one of the GameCenter links. You are looking for a resource with a '.json' extension, like the URL below:

    Request URL:http://www.nfl.com/liveupdate/game-center/2016091104/2016091104_gtd.json?random=1487543900000
    Request Method:GET
    Status Code:200 OK

So it looks like GameCenter gets its data from a JSON resource, like the one above. After looking at several GameCenter pages, we can see that NFL.com assigns a 10-digit unique ID for a game (which we'll call gsis_id).  Digits 1-8 are the date of the game in %Y%m%d format (such as '20160911') and digits 9-10 look like a game number, such as 00, 01, etc. 

We can use this detective work to write code that gets all of the GameCenter data from one week of an NFL season.

```python
from bs4 import BeautifulSoup
import requests

def get_gamecenter_json(season, week):
    '''
    Gets all of the gamecenter data from 1 week of the NFL season
    
    Arguments:
        season: int from 2009 to present season
        week: int from 1 to 17 (for regular season)
        
    Returns:
        Python data structure of parsed JSON
    '''
    gcs = []
    sb = get_scoreboard(season, week)
    for link in get_gamecenter_links(sb):
        r = requests.get(link)
        r.raise_for_status()
        gcs.append(r.json())
    return gcs
```

We can now get all of the passing plays in a game, which will be used to compute Air Yards.

```python
# needed for python 2-3 compatibility
from future.utils import iteritems

def passing_plays(gc):
    '''
    Filters out passing plays from all plays in the game
    
    Arguments:
        gc: dict that is parsed JSON from NFL.com

    Returns:
        pp: list of dicts of passing plays
    '''
    pp = []
    for drive_id, plays in iteritems(drives):
        for play_id, play in iteritems(plays):
            desc = play.get('desc', None)
            if desc and 'complete' in desc and 'pass' in desc: 
                pp.append({'drive_id': drive_id, 'play_id': play_id, 'play': play})   
    return pp
```

## Incomplete Air Yards

The data is different for incomplete and complete passes. For the former, incomplete air yards are found in statId 112 for the quarterback.

    {
        "sequence": 4,
        "clubcode": "KC",
        "playerName": "A.Smith",
        "statId": 112,
        "yards": 12
    }
    
Smith's pass would've gained 12 yards, but statId 112 doesn't say who the pass was thrown to. So we need to identify the receiver who is on the same team as the QB (in this case, J. Maclin). We can get Maclin's name either by looking at the 'desc' field or, more reliably, by finding the receiver in the players section.  The functions below show how we could identify the intended target of an incomplete pass.

```python
def find_receiver_by_desc(play):
    '''
    Finds the receiver on an incomplete pass play
    
    Arguments:
        play: dict of play information
        
    Returns:
        playerName: str such as A.Smith or J.Maclin
    '''
    desc = play.get('desc', None)
    if desc:
        name = desc.split(' ')[-1]
        for player_id, p in play.get('players').items():
            if player_id == '0':
                continue
            elif p.get('playerName') == name:
                return (player_id, name)   
    return None

def find_receiver_incomplete(play):
    '''
    Finds the receiver on an incomplete pass play
    
    Arguments:
        play: dict of play information
        
    Returns:
        player: tuple with 2 elements, player_id and name
    '''
    for player_id, p in play.get('players').items():
        if player_id == '0':
            continue
        for seq in p:
            if p.get('statId', 0) == 115 and p.get('yards', 1) == 0:
                return (player_id, p.get('playerName'))
```

We can now tally up all of the incomplete air yards by receiver.

```python
from collections import defaultdict

def sum_incomplete_air_yards(plays):
    '''
    Adds up incomplete air yards by playerId
    
    Arguments:
        plays: list of dict {'drive_id': drive_id, 'play_id': play_id, 'play': play}
    
    Returns:
        dict: player_id is key, incomplete air yards is value
    '''
    iay = defaultdict(int)
    
    # we aren't concerned about the play_id or drive_id keys
    for p in [play.get('play') for play in plays]:
        desc = p.get('desc', None)
        
        # make sure incomplete pass
        if 'incomplete' in desc:
            player_id, name = find_receiver_incomplete(p)
            
            # now get air yards
            # collapse per-player 'sequences' into single list
            # then find the incomplete air yards in statId 112 and increment iay
            for seq in [item for sublist in p.get('players') for item in sublist]:
                if seq.get('statId', None) == 112:
                    iay[player_id] += seq.get('yards')
                
    return iay
```

# Understanding the nfldb and nflgame libraries

If you have no interest in building a scraper and calculating Air Yards data yourself, do not fret. There is a pair of outstanding python libraries--nfldb and nflgame--which help you set up a postgresql database of all of the play-by-play data since 2009. It is beyond the scope of this article to help you setup the database. 

My advice is that it is worth the learning curve because you can do things (relatively) simply in postgresql that are insanely difficult or impossible to do in a spreadsheet or simpler database engine. I am not a professional programmer or database administrator and had no experience with postgresql until I wanted to install nfldb and nflgame.

    CREATE MATERIALIZED VIEW air_yards AS

    WITH qb AS (
     SELECT play_player.gsis_id,
        play_player.drive_id,
        play_player.play_id,
        play_player.team,
        play_player.passing_cmp,
        play_player.passing_att,
        play_player.passing_yds,
        play_player.passing_incmp_air_yds AS air_yds,
        play_player.player_id as qb
       FROM play_player
      WHERE passing_incmp_air_yds > 0 AND (play_player.gsis_id::text IN ( SELECT game.gsis_id
               FROM game
              WHERE game.season_type = 'Regular'::season_phase))
              AND passing_att > 0 AND passing_cmp = 0
    ),

    rcvr AS (
    SELECT qb.*, pp.receiver, pp.receiving_rec, pp.receiving_yds, pp.receiving_yac_yds
    FROM qb
    LEFT JOIN
    (SELECT gsis_id, drive_id, play_id, player_id, team, passing_att, receiving_rec, receiving_yds, receiving_yac_yds, player_id AS receiver FROM play_player WHERE passing_att = 0) AS pp
    ON qb.gsis_id = pp.gsis_id AND qb.drive_id = pp.drive_id AND qb.play_id = pp.play_id AND qb.team = pp.team
    ),

    icay AS (
    SELECT 
      rcvr.receiver as player_id, p2.full_name, rcvr.gsis_id, rcvr.drive_id, rcvr.play_id, rcvr.team, rcvr.receiving_rec, rcvr.air_yds, rcvr.receiving_yds, rcvr.receiving_yac_yds
    FROM
      rcvr
    LEFT JOIN
      player as p
    ON rcvr.qb = p.player_id
    LEFT JOIN
      player as p2
    ON rcvr.receiver = p2.player_id),

    cay AS (
    SELECT 
      pp.player_id, p.full_name, pp.gsis_id, pp.drive_id, pp.play_id, pp.team, pp.receiving_rec, (pp.receiving_yds - pp.receiving_yac_yds) as air_yds,
      receiving_yds, receiving_yac_yds
      FROM
      play_player AS pp
    LEFT JOIN
      player as p
    ON 
      pp.player_id = p.player_id
    WHERE
      pp.receiving_rec = 1
      AND (pp.gsis_id::text IN ( SELECT game.gsis_id
               FROM game
              WHERE game.season_type = 'Regular'::season_phase))
    ),

    comb AS (
    SELECT * FROM cay
    UNION ALL
    SELECT * FROM icay)

    SELECT g.season_year, comb.*
    FROM comb
    JOIN game as g
    ON comb.gsis_id = g.gsis_id

    WITH DATA;

    ALTER TABLE public.air_yards
      OWNER TO nfldb;

    CREATE INDEX on public.air_yards (player_id, air_yds, receiving_rec);
    CREATE INDEX on public.air_yards (player_id, receiving_rec, season_year);
    CREATE INDEX on public.air_yards (player_id, receiving_yac_yds, season_year);

```python
from collections import defaultdict
import time

import requests

def get_scoreboards(delay=2):
    '''
    Gets all scoreboard pages since 2009 for regular season
    
    Arguments:
        delay: int in seconds (be polite - don't flood sites with requests)
        
    Returns:
        dict with years as keys, dict as values with week as key and HTML content as value
    '''
    base_url = 'http://www.nfl.com/scores/{season}/REG{week}'

    # default dict will add dictionary as value for each season
    # will also keep keys sorted in the order they are added
    scoreboards = defaultdict(dict)
    
    # loop through seasons and weeks
    for season in range(2009, 2017):
        for week in range(1,18):
            r = requests.get(base_url.format(season, week))
            r.raise_for_status()
            scoreboards[season][week] = r.content
            time.sleep(delay)
    return scoreboards
```
